```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data_after_feature_engineering_2.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Total_Stops',y='Price',data=df)
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='Airline_Original', y='Price', data=df)
plt.xticks(rotation=90)
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='Source_Original', y='Price', data=df,)
plt.xticks(rotation=90)
plt.title("Price Distribution by Source City")
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='Source_Original', y='Price', data=df,)

plt.title("Price Distribution by Source City")
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='Destination_Original', y='Price', data=df)
plt.title("Price Distribution by Destination City")
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x = 'Dep_hour',y = 'Price', data=df)
plt.title("Flight Price by Departure Hour")
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Total_Stops', y='Price', data=df)
plt.title("Effect of Number of Stops on Price")
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Total_Stops', y='Price', data=df)
plt.title("Effect of Number of Stops on Price")
```

```python
df['Total_Duration'] = df['Duration_hours'] * 60 + df['Duration_mins']
plt.figure(figsize=(10,5))
sns.scatterplot(x='Total_Duration', y='Price', data=df)
plt.title("Duration (mins) vs Flight Price")
```

```python
plt.figure(figsize=(12,5))
sns.countplot(x='Airline_Original', data=df, order=df['Airline_Original'].value_counts().index)
plt.xticks(rotation=90)
plt.title("Flight Count per Airline")
```
