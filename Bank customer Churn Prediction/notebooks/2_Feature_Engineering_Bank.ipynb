```python
import pandas as pd
```

```python
df= pd.read_csv('../data/Churn_Modelling.csv')
```

```python
df.head(10)
```

```python
# 1. CLV Estimation
df['MonthlyProfit'] = df['Balance'] * 0.005  
df['CLV'] = df['MonthlyProfit'] * (df['Tenure'] * 12)
```

```python
# 2. Behavioral Features
df['ActivityScore'] = df['IsActiveMember'] * 2 + df['HasCrCard']
df['BalanceToSalaryRatio'] = df['Balance'] / (df['EstimatedSalary'] + 1)
```

```python
# 3. Tenure Bins
df['TenureGroup'] = pd.cut(df['Tenure'], bins=[0, 2, 5, 10], labels=['New', 'Mid', 'Loyal'])
```

```python
features = [
    'CreditScore', 'Age', 'Tenure', 'Balance', 
    'NumOfProducts', 'HasCrCard', 'IsActiveMember',
    'BalanceToSalaryRatio', 'CLV', 'Geography', 'Gender'
]
```

```python
X = pd.get_dummies(df[features], drop_first=True)
y = df['Exited']
```

```python
X.to_csv("../data/bank_features.csv", index=False)
```
